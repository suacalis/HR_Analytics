```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```
iky= pd.read_csv('ikyogrenci2019.csv', sep=';')
```

```
iky.head()
```

```
iky.info()
```

```
for col in iky.columns:
    print(col, iky[col].nunique())
```

```
# creating a list of the categorical variables
cat_var = ['Program',
'OgretimUyesi',
'GirisYili']
```

```
for col in cat_var:
    print(iky.groupby(col)[col].count(), '\n')
```

```
sns.catplot(x = 'GirisYili', hue = 'OgretimUyesi', data = iky, kind = 'count')
plt.show()
```

```
sns.catplot(x = 'GirisYili', hue = 'Program', data = iky, kind = 'count')
plt.show()
```

```
px.histogram(iky, x="GirisYili", y="GirisYili", histfunc="sum", color="Program")
```
